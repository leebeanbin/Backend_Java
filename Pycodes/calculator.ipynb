```python
import math

fMean = 37.31
fDis  = 2.228
fMeanDis = 0.1185
nNum  = math.sqrt(10)


res = fMean-(fDis*(fMeanDis/nNum))
res2 = fMean+(fDis*(fMeanDis/nNum))
res3 = res2 - res
print("%.5f %.5f %.5f" % (res, res2, res3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 입력한 10개의 체온 결과 (예시)
temperature_data = [37.3, 37.4, 37.3, 37.3, 37.3, 37.3, 37.4, 36.9, 37.2, 37.2]

# 평균과 표준 편차 계산
mean_temperature = np.mean(temperature_data)
std_deviation = np.std(temperature_data)

# 정규 분포 그래프 데이터 생성
x = np.linspace(mean_temperature - 3 * std_deviation, mean_temperature + 3 * std_deviation, 100)
y = (1 / (std_deviation * np.sqrt(2 * np.pi))) * np.exp(-(x - mean_temperature) ** 2 / (2 * std_deviation ** 2))

# 그래프 그리기
plt.plot(x, y, label='Normal Distribution')
plt.xlabel('Temperature')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Temperature')
plt.legend()
plt.grid(True)

# 그래프 표시
plt.show()
```
